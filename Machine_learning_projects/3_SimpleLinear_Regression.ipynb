```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('marks.csv')
```

```python
plt.scatter(df.SAT,df.GPA)
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
```

```python
regressor.fit(df[['SAT']],df.GPA)
```

```python
regressor.predict([[1943]])
```

```python
regressor.coef_
```

```python
plt.scatter(df.SAT,df.GPA)
plt.plot(df.SAT,regressor.predict(df[['SAT']]))
plt.show()
```

import libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

create dataframe 

```python
df=pd.read_csv("house_price.csv")
df
```

```python
df.isna().sum()
```

```python
bed=df.Bedroom.median()
bed
```

```python
import math
bed_room=math.floor(bed)
bed_room
```

```python
df.Bedroom=df.Bedroom.fillna(bed_room)
df
```

Create Model

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
```

fit the model

```python
regressor.fit(df[['Areas','Bedroom','Years']],df.Price)
```

Predict the model

```python
regressor.predict([[4500,4,12]])
```

```python
regressor.coef_
```

```python
regressor.intercept_
```

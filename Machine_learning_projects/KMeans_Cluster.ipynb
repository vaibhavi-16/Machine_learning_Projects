```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Mall_Customers.csv')
df
```

```python
plt.scatter(df['Annual Income (k$)'] , df['Spending Score (1-100)'])
plt.show()
```

```python
from sklearn.cluster import KMeans
km = KMeans(n_clusters = 5)
km
```

```python
y_pred = km.fit_predict(df[['Annual Income (k$)' , 'Spending Score (1-100)']])
y_pred
```

```python
df['Cluster'] = y_pred
df
```

```python

df1 = df[df.Cluster == 0]
df2 = df[df.Cluster == 1]
df3 = df[df.Cluster == 2]
df4 = df[df.Cluster == 3]
df5 = df[df.Cluster == 4]
```

```python

plt.scatter(df1['Annual Income (k$)'] ,  df1['Spending Score (1-100)'] , color = 'green' )
plt.scatter(df2['Annual Income (k$)'] ,  df2['Spending Score (1-100)'] , color = 'yellow')
plt.scatter(df3['Annual Income (k$)'] ,  df3['Spending Score (1-100)'] , color = 'red')
plt.scatter(df4['Annual Income (k$)'] ,  df4['Spending Score (1-100)'] , color = 'pink')
plt.scatter(df5['Annual Income (k$)'] ,  df5['Spending Score (1-100)'] , color = 'brown')
plt.show()
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
```

```python
scaler.fit(df[['Annual Income (k$)']])
df['Annual Income (k$)'] = scaler.transform(df[['Annual Income (k$)']])

scaler.fit(df[['Spending Score (1-100)']])
df['Spending Score (1-100)'] = scaler.transform(df[['Spending Score (1-100)']])

df
```

```python
from sklearn.cluster import KMeans
km = KMeans(n_clusters=5)
```

```python
y_pred = km.fit_predict(df[['Annual Income (k$)' , 'Spending Score (1-100)']])
y_pred
```

```python
km.cluster_centers_
```

```python
k_rng = range(1,10)
see = []
for k in k_rng:
  km = KMeans(n_clusters=k)
  km.fit(df[['Annual Income (k$)' , 'Spending Score (1-100)']])
  see.append(km.inertia_)

plt.plot(k_rng,see)
plt.show()
```

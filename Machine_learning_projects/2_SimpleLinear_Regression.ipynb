```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('Salary_Data.csv')
```

```python
plt.scatter(df.YearsExperience,df.Salary,color='red')
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
obj=LinearRegression()
```

```python
obj.fit(df[['YearsExperience']],df.Salary)
```

```python
obj.predict([[4.0]])
```

```python
obj.coef_
```

```python
plt.scatter(df.YearsExperience,df.Salary,color='red')
plt.plot(df.YearsExperience,obj.predict(df[['YearsExperience']]))
plt.show()
```

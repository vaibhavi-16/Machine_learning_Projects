```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('salary_company.csv')
df
```

```python
x_n = df.drop('salary_more_than_100k' , axis='columns')
y = df['salary_more_than_100k']
print(x_n)
```

```python
from sklearn.preprocessing import LabelEncoder
obj = LabelEncoder()
```

```python
x_n['company_name'] = obj.fit_transform(x_n['company'])
x_n['job_name'] = obj.fit_transform(x_n['job'])
x_n['degree_name'] = obj.fit_transform(x_n['degree'])
print(x_n)
```

```python
x = x_n.drop(['company','job','degree'], axis= 'columns')
print(x)
```

```python
print(y)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.20,random_state=0)
```

```python
from sklearn.tree import DecisionTreeClassifier
classi = DecisionTreeClassifier()
```

```python
classi.fit(x_train,y_train)
```

```python
y_pred = classi.predict(x_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

Import dataset

```python
df = pd.read_csv('price.csv')
df
```

Create graph

```python
plt.xlabel("Price of area")
plt.ylabel("area as sqrt. feet")
plt.scatter(df.Area,df.Price)
plt.title("Graph of Linear Regression")
plt.show()
```

Create Model

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
```

Train Model

```python
regressor.fit(df[['Area']],df.Price)
```

Predict the model

```python
regressor.predict([[4500]])
```

predict coefficent

```python
regressor.coef_
```

Predict intercept

```python
regressor.intercept_
```

Draw a best fit line

```python
plt.xlabel("Price of area")
plt.ylabel("area as sqrt. feet")
plt.scatter(df.Area,df.Price)
plt.plot(df.Area,regressor.predict(df[['Area']]))
plt.title("Graph of Linear Regression")
plt.show()
```

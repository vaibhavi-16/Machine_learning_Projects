```python
plt.scatter(df0['sepal length (cm)'], df0['sepal width (cm)'], color='blue')
plt.scatter(df1['sepal length (cm)'], df1['sepal width (cm)'], color='pink')
plt.show()
```

```python
plt.scatter(df0['petal length (cm)'], df0['petal width (cm)'], color='blue')
plt.scatter(df1['petal length (cm)'], df1['petal width (cm)'], color='pink')
plt.show()
```

```python
x = df.iloc[:,[0,1,2,3]].values
y = df.iloc[:,-2].values
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test= train_test_split(x,y,test_size=0.25,random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
model = StandardScaler()
x_train = model.fit_transform(x_train)
x_test = model.transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression  
regressor = LogisticRegression()
```

```python
regressor.fit(x_train,y_train)
```

```python
y_pred = regressor.predict(x_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```
